```python
import pickle

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
plt.style.use('seaborn-darkgrid')
```

```python
RESAMPLE = '1d'
```

```python
price = pickle.load(open('price_data.pkl', 'rb'))
```

```python
# price = price.fillna
# price = price.dropna()
```

```python
price = price.resample(RESAMPLE).mean()
price = price.fillna(method='ffill')
price = price.dropna()
price.head()
```

```python
tokens = price.columns.tolist()
tokens
```

```python
initial_price = price.iloc[0]
```

```python
initial_weights = np.ones(len(tokens)) / len(tokens)
initial_weights
```

```python
INVEST = 1000
```

```python
initial_allocation = INVEST * initial_weights / initial_price
initial_allocation
```

```python
initial_positions = initial_allocation * initial_price
initial_positions
```

```python
initial_portfolio_value = initial_positions.sum().sum()
initial_portfolio_value
```

```python
hodl = {}
rb = {}

current_allocation = initial_allocation

for i, current_price in price.iterrows():
    # hodl positions
    hodl[i] = initial_allocation * current_price
    
    # current positions
    current_positions = current_allocation * current_price
    rb[i] = current_positions
    
    # rebalance
    current_portfolio_value = current_positions.sum()
    current_allocation = current_portfolio_value * initial_weights / current_price
    
hodl = pd.DataFrame(hodl).T
rb = pd.DataFrame(rb).T
```

```python
hodl_value = hodl.sum(axis=1)
rb_value = rb.sum(axis=1)

hodl_perf = 1 - (1 / ((hodl_value.iloc[-1] / hodl_value.iloc[0])))
rb_perf = 1 - (1 / ((rb_value.iloc[-1] / rb_value.iloc[0])))

fig, ax = plt.subplots(1, figsize=(16, 9))
ax.plot(hodl_value, label='hodl {:.1f}%'.format(hodl_perf * 100), linewidth=3)
ax.plot(rb_value, label='snapfund {:.1f}%'.format(rb_perf * 100), linewidth=3)
ax.grid(True, 'both')
ax.set_ylabel('portfolio value [%]', fontsize=14)
ax.legend(fontsize=16)
ax.set_title('snapfund | periodic rebalancing', fontsize=18)
fig.autofmt_xdate();
```

# Threshold rebalancing

```python
THR = 0.2
```

```python
hodl = {}
rb = {}

current_allocation = initial_allocation

for i, current_price in price.iterrows():
    # hodl positions
    hodl[i] = initial_allocation * current_price
    
    # current positions
    current_positions = current_allocation * current_price
    rb[i] = current_positions
    
    # rebalance
    current_portfolio_value = current_positions.sum()
    current_weights = current_positions / current_portfolio_value
    weights_diff = np.abs(current_weights - initial_weights)
    if any(weights_diff > THR):
        print(i, 'rebalancing')
        current_allocation = current_portfolio_value * initial_weights / current_price

hodl = pd.DataFrame(hodl).T
rb = pd.DataFrame(rb).T
```

```python
hodl_value = hodl.sum(axis=1)
rb_value = rb.sum(axis=1)

hodl_perf = 1 - (1 / ((hodl_value.iloc[-1] / hodl_value.iloc[0])))
rb_perf = 1 - (1 / ((rb_value.iloc[-1] / rb_value.iloc[0])))

fig, ax = plt.subplots(1, figsize=(16, 9))
ax.plot(hodl_value, label='hodl {:.1f}%'.format(hodl_perf * 100), linewidth=3)
ax.plot(rb_value, label='snapfund {:.1f}%'.format(rb_perf * 100), linewidth=3)
ax.grid(True, 'both')
ax.set_ylabel('portfolio value [%]', fontsize=14)
ax.legend(fontsize=16)
ax.set_title('snapfund | threshold rebalancing {}%'.format(int(THR * 100)), fontsize=18)
fig.autofmt_xdate();
```
